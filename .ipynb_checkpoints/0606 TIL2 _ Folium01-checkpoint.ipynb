# 시각화 하는 방법

###  지도 생성 = folium.Map
- m = folium.Map(location = [위도, 경도], zoom_start = 확대정도)

### 정보 추가  =  .add_to(m)
- 마커추가
    - folium.Marker(location = [위도, 경도]).add_to(m)
- 원추가
    - folium.CircleMarker(location = [위도, 경도]), radius = 원의 크기).add_to(m)
- 추가옵션
    - tooltip = '마우스 올렸을 때 보여지는 정보'
    - popup = '클릭시 보여지는 정보 ' 
- etc
    - clickForMarker('체크').add_to(m): 원마커 추가


# 지도 생성

```python
# 서울역 위도와 경도 : [37.5536067, 126.9674308]
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 15)
m
```

# 지도에 정보 추가하기

```python
#마커 추가 1 
folium.Marker([37.5536067, 126.9674308]).add_to(m)
m
```

```python
# 마커 추가 2 _ tooltip_ 마우스 올려서 보이게할 정보
folium.Marker([37.5536067, 126.9674308], 
              tooltip = '서울역').add_to(m)
m
```

```python
# 마커 추가 2 _ popup_ 클릭시 올려서 보이게할 정보
folium.Marker([37.5536067, 126.9674308], 
              popup = '서울역').add_to(m)
m
```

```python
#CircleMarker _ 원마커 추가
folium.CircleMarker( [37.5536067, 126.9674308],
                  radius = 20,
                  tooltip = '서울역').add_to(m)
m
```

# Minimap 추가

```python
from folium.plugins import MiniMap
```

```python
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 12)
```
