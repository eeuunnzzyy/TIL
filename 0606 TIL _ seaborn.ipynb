# 수치형 * 수치형 데이터 그래프

## 정형
- scatterplot
- lmplot
- pairplot

## 비정형 (변화가능)
- relplot
- jointplot은 수치 * 수치 데이터

```python
sns.scatterplot(data = raw, x= 'total_bill', y='tip', hue = 'sex')
```

```python
sns.relplot(data = raw, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(
    data = raw,
    x = 'total_bill', y = 'tip',
    kind = 'line')
```

```python
sns.lmplot(data = raw, x = 'total_bill', y= 'tip', hue = 'smoker')
```

```python
# pairplot은 데이터에 들어있는 수치형데이터만을 똑 꺼내서 전체를 다 보여준다.
sns.pairplot(
    data = raw
)
```

```python
sns.pairplot(data =raw, hue = 'sex')
```

```python
#jointplot
#각각의 축에 대한 히스토그램과 두개를 합친 scatterplot 합쳐짐
sns.jointplot(
    data = raw,
    x = 'tip', y = 'total_bill'
)
```

```python
sns.jointplot(
    data = raw,
    x = 'tip', y = 'total_bill', 
    kind = 'kde'
)
```

```python
sns.jointplot(
    data = raw,
    x = 'tip', y = 'total_bill',
    kind = 'hex'
)
```

# 수치형 * 카테고리형 그래프 

- barplot, boxplot 은 수치형 * 카테고리형 데이터관의 분포 보여줌


```python
sns.barplot(
    data = raw,
    x = 'day', y = 'tip'
)
```

```python
sns.boxplot(
    data = raw,
    x = 'day', y= 'tip'
)


#boxplot은 특정 기준에 데이터의 분포를 알기에 가장 좋은 그래프다.
#하지만 boxplot은 데이터의 분포는 표현하지만 해당하는 데이터의 개수는 표현하지 않는다.
#그래서 class마다 데이터의 개수가 상이할 때는, 적을땐 이 값을 그대로 해석해서 받아들이기에는 문제가 생긴다
```

```python
#그래서 swarmplot으로 데이터의 개수도 같이 표현해서 분포를 표현해주는 것이다.

sns.swarmplot(
    data = raw,
    x = 'day', y = 'tip',
    hue = 'smoker'
)
```

```python
#hue를 기준으로 나눈 데이터 끼리 안겹쳐 보이게 하려면 dodge = True
#dodge의 한국어 뜻은 피하다.
sns.swarmplot(
    data = raw,
    x = 'day', y = 'tip',
    hue = 'smoker',
    dodge = True
)
```

```python
#동시에 보이게 하기
sns.boxplot(
    data = raw, x = 'day', y = 'tip',
    hue = 'smoker', dodge = True)
sns.swarmplot(
    data = raw, x = 'day', y = 'tip',
    hue = 'smoker', dodge = True)
```

# 히트맵

```python
#히트맵은 두 카테고리형 데이터 분포에 대한 수치형 데이터의 값을 한눈에 볼 수 있다.
#히트맵 이용하기 전에, 두 카테고리형 데이터 분포에 대한 수치형 데이터의 값을 피벗테이블로 만들어줘야함
```

```python
df = raw.pivot_table(
    index = 'day', columns = 'size',
    values = 'tip', aggfunc = 'mean')
df
```

```python
sns.heatmap(df, 
            annot = True, 
            fmt = '0.0f',
            cmap = 'Blues'
           )
```
