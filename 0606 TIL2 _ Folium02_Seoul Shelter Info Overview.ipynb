# Grouping 

## ClusterMarker

```python
from folium.plugins import MarkerCluster
```

```python
m = folium.Map(location = [37.5536067, 126.9674308],
              zoom_start=12)


#########makrer_cluster는 일단 map하나 만들어놓고 지정#####
marker_cluster = MarkerCluster().add_to(m)


#######여기에 입혀#####
#######add_to(marker_cluster)로!!
for i in range(len(raw)):
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long], tooltip = name).add_to(marker_cluster)
    
m
```

```python
from folium.plugins import MiniMap
```

```python
m = folium.Map(location = [37.5536067, 126.9674308],
              zoom_start=12)


#########makrer_cluster는 일단 map하나 만들어놓고 지정#####
marker_cluster = MarkerCluster().add_to(m)

#####minimap 추가
minimap = MiniMap()
minimap.add_to(m)

#######마커 클러스터여기에 입혀#####
#######add_to(marker_cluster)로!!
for i in range(len(raw)):
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long], tooltip = name).add_to(marker_cluster)
    
m
```
